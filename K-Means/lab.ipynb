```python
reshaped_image = new_image.reshape(-1,3)
```

```python
reshaped_image.shape
```

```python
reshaped_image[5]
```

```python
from sklearn.cluster import KMeans
```

```python
k_mean = KMeans(n_clusters=3, random_state=43)
```

```python
k_mean.fit(reshaped_image)
```

```python
value_k_means = k_mean.predict(reshaped_image)
```

```python
value_k_means
```

```python
(value_k_means == 2).sum()
```

```python
number_R, count_R = np.unique(reshaped_image[:,0], return_counts=True)
number_G, count_G = np.unique(reshaped_image[:,1], return_counts=True)
number_B, count_B = np.unique(reshaped_image[:,2], return_counts=True)
```

```python
plt.bar(number_R,count_R)
plt.show()
```

```python
plt.bar(number_G,count_G)
plt.show()
```

```python
plt.bar(number_B,count_B)
plt.show()
```

```python
unique_values, counts = np.unique(value_k_means, return_counts=True)


for value, count in zip(unique_values, counts):
    print(f"{value} occurs {count} times")
```
